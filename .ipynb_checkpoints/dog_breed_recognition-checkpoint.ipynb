## Downloading images from the Internet

```python
def search_images(term, max_images=200): 
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

## Dog test photo downloading and displaying
Downloading a picture of the dog, which later will be used to test the trained model and calculate prediction of belonging to a given breed

```python
from fastdownload import download_url
from fastai.vision.all import *

dest = 'dog.jpg'

download_url(
    search_images('dog photos', max_images=1)[0],
    dest,
    show_progress=False
)

Image.open(dest).to_thumb(256, 256)
```

## Downloading photos of dogs of selected breeds and dividing them into folders
According to the internet, the 6 most popular dog breeds are labrador retriever, german shepherd, golden retriever, bolldog, beagle and french bulldog. Here is downloading of their photos and dividing them into directories with corresponding breed names. Also searching for photos captured at different moments of movement (walking, running).

```python
from time import sleep

breeds = ['Labrador Retriever', 'German Shepherd', 'Golden Retriever', 'Bulldog', 'Beagle', 'French Bulldog']
path = Path('DogBreeds')

for breed in breeds:
    dest = (path/breed)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{breed} photo'))
    sleep(5)
    download_images(dest, urls=search_images(f'walking {breed} photo'))
    sleep(5)
    download_images(dest, urls=search_images(f'running {breed} photo'))
    sleep(5)
    resize_images(path/breed, max_size=400, dest=path/breed)
```

## Removing non verified images not to cause failures

```python
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
len(failed)
```

## DataLoader - training and validation sets
Creating a training set (the images used to create a model) and a validation set (the images used to check the accuracy of a model -- not used during training). 
Showing some downloaded pictures.

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=32)

dls.show_batch(max_n=6)
```

## Training model with resnet18

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(6)
```

## Displaying probability for each breed as bar graph
Combining breeds and probabilities by index, sorting descending and displaying as a bar graph with values on bars and breed names as x-axis.

```python
import matplotlib.pyplot as plt

def probability_bar_graph(probs):
    combined_data = list(zip(breeds, probs.tolist()))
    sorted_data = sorted(combined_data, key=lambda x: x[1], reverse=True)
    sorted_breeds = [data[0] for data in sorted_data]
    sorted_tensor = [data[1] for data in sorted_data]
    plt.bar(range(len(sorted_breeds)), sorted_tensor)

    # adding values on bars
    for i, value in enumerate(sorted_tensor):
        plt.text(i, value, f'{value*100:.2f}%', ha='center', va='bottom')
    plt.xticks(range(len(sorted_breeds)), sorted_breeds, rotation='vertical')

    plt.xlabel('Breeds')
    plt.ylabel('Tensor Value [prob.]')
    plt.title('Breed probabilities')

    # print(sorted_data)
    plt.show()
```

## Results
Printing class with the highest probability and displaying the graph with the probability for all breeds.

```python
breed, tensor, probs = learn.predict(PILImage.create('dog.jpg'))
print(f"[prob.: {probs[tensor.item()]*100:.2f}%]: {breed} recognized")

probability_bar_graph(probs)
```
